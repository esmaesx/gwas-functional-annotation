# GWAS → Multi-omics Prioritization — Overview

A transparent pipeline that connects GWAS signals to transcriptomics/proteomics for target ranking.

**Scope:**
- Load GWAS summary stats
- Add eQTL/pQTL layers
- Score and rank targets with interpretable logic

**Methods (high-level):**
- Normalize GWAS signals
- Integrate eQTL/pQTL evidence
- Rank targets with interpretable scoring

Uses only public data.
